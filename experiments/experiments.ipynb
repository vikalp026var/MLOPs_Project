```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df
```

```python
X=df.drop(labels=['price'],axis=1)
y=df['price']
```

```python
X
```

```python
df['color'].value_counts()
print("=================")
df['clarity'].value_counts()
```

```python
df.drop(labels=['Unnamed: 0'],axis=1,inplace=True)
```

```python
cut_categories=['Fair','Good','Very Good','Premium','Ideal']
color_categories=['D','E','F','G','H','I','J']
calrity_categories=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
num_pipeline=Pipeline(
     steps=[
          ("imputer",SimpleImputer()),
          ("scaler",StandardScaler())
          
     ]
)
```

```python
cat_pipeline=Pipeline(
     steps=[
          ("imputer",SimpleImputer(strategy="most_frequent")),
          ("ordinalencoder",OrdinalEncoder(categories=[cut_categories,color_categories,calrity_categories]))
     ]
)
```

```python
preprocessor=ColumnTransformer(
     [
          ("num_pipeline",num_pipeline,df[num_caolumn].columns.drop(labels=['price'])),
          ("categorical_pipeline",cat_pipeline,df[categorical_column].columns)
     ]
)
```

```python
preprocessor
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
```

```python
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
from xgboost import XGBRegressor 
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models={
     'LinearRegression':LinearRegression(),
     'Lasso':Lasso(),
     'Ridge':Ridge(),
     'Elasticnet':ElasticNet(),
     'RandomForest':RandomForestRegressor(),
     'xgboost':XGBRegressor()
}
```

```python
train_model_list=[]
model_list=[]
r2_list=[]
```

```python
list(models)
```

```python
def evaluate_model(true,pred):
     r2_scor=r2_score(true,pred)
     mse=mean_squared_error(true,pred)
     mae=mean_absolute_error(true,pred)
     return r2_scor,mse,mae
```

```python
for model_name,model in models.items():
     model.fit(X_train,y_train)
     y_pred=model.predict(X_test)
     print(f"model which is fitted is : {model_name}")
     r2_scor,mse,mae=evaluate_model(y_test,y_pred)
     r2_list.append(r2_scor)
     print(f" MSE of {model_name} is :{mse}")
     print(f"mae of {model_name} is :{mae}")
     print(f"r2_score of {model_name} is {r2_scor}")
     print("="*40)
     print("\n")
     
```
